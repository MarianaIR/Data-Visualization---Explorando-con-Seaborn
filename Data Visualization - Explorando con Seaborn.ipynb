Traduciendo Variables


```python
#Traducimos nombres de nuestras Variable al español, primero
#empezamos con los nombres de las COLUMNAS, visualizamos los nombres
datos.columns
```

```python
#Ahora usamos un DICCIONARIO para cambiar los nombres al español
dic_columnas = {'LIMIT_BAL':"limite",
                'CHECKING_ACCOUNT':"cuenta_corriente",
                'EDUCATION':"escolaridad",
                'MARRIAGE':"estado_civil",
                'AGE':"edad",
                'BILL_AMT':"valor_factura",
                'PAY_AMT':"valor_pago",
                'DEFAULT':"moroso" }
```

```python
#Una vez traducido , creamos un nuevo DATAFRAME y lo llamamos
#TARJETAS
tarjetas = datos.rename(columns = dic_columnas)
```

```python
#Ejecutamos el nuevo DATAFRAME TARJETAS modificado
tarjetas.head()
```

```python
#Ahora camabiamos los nombres de nuestras Variables de los columnas, comenzamos
#la Variable de CUENTA_CORRIENTE, primero vemos que valores tenemos disponible
#en esta columna, usamos metodo .UNIQUE, nos trae los valores de esta columna
tarjetas.cuenta_corriente.unique()
```

```python
#De la misma forma que hacemos con las columnas, vamos a usar tambien un
#DICCIONARIO
dic_cuenta = {"Yes": "Si", "No": "No"}
```

```python
#Ahora cambiamos la Variable con el metodo .MAP, y pasamos como parametro
#el DICCIONARIO
tarjetas.cuenta_corriente.map(dic_cuenta)
```

```python
#Luego atribuimos lo que hicimos a nuestra base de datos
tarjetas.cuenta_corriente = tarjetas.cuenta_corriente.map(dic_cuenta)
tarjetas.head()
```

```python
#Luego hacemos lo mismo con la columna ESCOLARIDAD
tarjetas.escolaridad.unique()
```

```python
#DICCIONARIO
dic_escolaridad = {'2.University' : "2.Universidad",
                   '3.Graduate School': "3.Pos-graduacion",
                   '1.High School' : "1.Colegio"}
```

```python
#FUNCION
tarjetas.escolaridad = tarjetas.escolaridad.map(dic_escolaridad)
tarjetas.head()
```

```python
#Hacemos lo mismo con la columna ESTADO_CIVIL
tarjetas.estado_civil.unique()
```

```python
#DICCIONARIO
dic_estado_civil = {'Married': "Casado/a",
                    'Single': "Soltero/a" }
```

```python
#Creamos COLUMNA y LISTAS
bins = [20,30,40,50,100] #(lista edad)
nombres = ["20-30", "30-40", "40-50", "50+"] #(Rando de Grupos)
```

```python
#Ahora creamos nuestra COLUMNA en nuestra base de datos
#(tarjetas["edad"] = columna que vamos a agrupar, bins = cortes de rango
#labels = nombres = nombres de rangos
tarjetas["rango_edad"] = pd.cut(tarjetas["edad"], bins, labels = nombres)
tarjetas.head() #Ejecutamos, crea columna RANGO_EDAD y la relaciona con EDAD
```

```python
#EJEMPLO AULA -->
#Una persona analiza nuestra base de datos y considera que deberíamos agrupar
#la columna de iu, porque considera que más adelante podríamos cruzarla
#con otras informaciones de forma más eficiente.
#Nos recomienda separarla en 4 grupos, como podríamos hacerlo:
bins = [0, 0.25, 0.50, 0.75, 1]
nombres = ['0%-25%', '25%-50%', '50%-75%', '75%-100%']

tarjetas['rango_iu'] = pd.cut(tarjetas['iu'], bins, labels=nombres)
tarjetas.head()
#Primero creamos una lista con los valores que serán
#nuestro punto de corte. Después creamos una lista con los nombres que le
#daremos a nuestros rangos y por últimos usaremos la función cut de pandas
#para crear nuestra variable en la base de datos.
```

Test de Hipotesis

```python
#Copiamos el grafico que creamos arriba, algo que nos permite este grafico de
#REGRESION es que podemos abrirlo para comparar al mismo tiempo con alguna
#columna que tengamos en este caso MOROSO
sns.lmplot(x = "iu", y = "valor_factura", data = tarjetas, col = "moroso");
#Podemos ver por separado la linea de REGRESION para cada uno, MOROSO y NO
#MOROSOS, pero tienen INCLINACION diferente en los 2 casos
```

```python
#Ahora vemos si la diferencia de INCLINACION en cada uno, es significativamente
#estadisticamente, para esto usamos la biblioteca SCIPY (RANKSUMS)
from scipy.stats import ranksums
#Vamos a escribir muestra 2 HIPOTESIS, en LINEA de TEXTO escribimos las 2 -->
```

H<sup>null<sup>
La distribucion de los grupos MOROSOS y NO MOROSOS es la misma

H<sup>alt<sup>
La distribucion de los grupos MOROSOS y NO MOROSOS, NO es la misma

```python
#Ahora usamos nuestra Variable RANKSUMS, primero vamos a crear los grupos MOROSOS
# y NO MOROSOS y para eso vamos a usar una QUERY
moroso = tarjetas.query("moroso == 1"). valor_factura
moroso
#Nos devuelve el Valor de Factura de nuestras perosonas MOROSAS (117)
```

```python
no_moroso = tarjetas.query("moroso == 0"). valor_factura
no_moroso
#Nos devuelve el Valor de Factura de nuestras perosonas MOROSAS (1383)
```

```python
#Ahora vamos a ver si la Distribucion de MOROSOS y NO_MOROSOS, es
#estadisticamente significativa, usamos funcion RANKSUMS
ranksums(moroso, no_moroso)
#El P_VALUE es del 1%, generalmente los valores abajo de 5% nos confirma
# la HIPOTESIS 2, NO es la misma distribucion de los MOROSOS y NO_MOROSOS
```

```python
#Guardamos el RESULTADO, en una Variable Resultado
resultado = ranksums(moroso, no_moroso)
print("El P_VALUE de nuestro test es {}".format(resultado.pvalue))
```
